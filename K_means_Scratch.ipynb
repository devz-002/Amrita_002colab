```python
#declaring clusters 3
k = 3
#declaring centroids 
c1 = data.iloc[0] 
c2 = data.iloc[int(len(data)/2-1)] 
c3 = data.iloc[int(len(data)-1)] 
```

```python
def CalcDist(Kdata):
    list1=[]
    list2=[]
    list3=[]
    for c in range(len(data)):
        list1.append(np.linalg.norm(c1-data.iloc[c][:2]))
    for c in range(len(data)):
        list2.append(np.linalg.norm(c2-data.iloc[c][:2]))
    for c in range(len(data)):
        list3.append(np.linalg.norm(c3-data.iloc[c][:2]))
    data["list1"] = list1
    data["list2"] = list2
    data["list3"] = list3
```

```python
def clust(dt):
    clust = []
    clus=[]
    for i in range(len(dt)):
        if data.iloc[i,2]< data.iloc[i,3] and data.iloc[i,4] :
            clus.append(1)
        elif data.iloc[i,3]< data.iloc[i,2] and data.iloc[i,4]:
            clus.append(2)
        else :
            clus.append(3)
    data["clust"] = clus
```
